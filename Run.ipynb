```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
```

```python
my_model = tf.keras.models.load_model('{DIRECTORY}/my_model.h5')
```

```python
img = cv2.imread('/content/OIP.jpg')
plt.imshow(img)
```

```python
resize = tf.image.resize(img,(256,256))
y = my_model.predict(np.expand_dims(resize/255,0))
```

```python
if y>0.5 :
  print('Tumour Present')
else:
  print('No tumour Present')
```
